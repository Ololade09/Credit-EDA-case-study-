Dealing with missing values in the dataset

```python
# Impute missing values for numerical columns with the mean
numeric_columns = ['AMT_ANNUITY', 'AMT_DOWN_PAYMENT', 'AMT_GOODS_PRICE', 'AMT_CREDIT', 'CNT_PAYMENT']
for col in numeric_columns:
    data[col].fillna(data[col].mean(), inplace=True)
```

```python
# For categorical column 'NAME_TYPE_SUITE', fill missing values with the most frequent category
data['NAME_TYPE_SUITE'].fillna(data['NAME_TYPE_SUITE'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
mean_rate_down_payment = data['RATE_DOWN_PAYMENT'].mean()
data['RATE_DOWN_PAYMENT'].fillna(mean_rate_down_payment, inplace=True)
```
